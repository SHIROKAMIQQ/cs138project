```python
from pyrosm import OSM
import pandas as pd

osm = OSM("map.osm.pbf")

buildings: pd.DataFrame = osm.get_buildings()
buildings.plot()
```

```python
import pandas as pd

buildings_coords = []

for index, row, in buildings.iterrows():
    coords = row['geometry'].exterior.coords
    for x, y in coords:
        buildings_coords.append((index, x, y))
    
df_buildings_coords = pd.DataFrame(buildings_coords, columns=['Name', 'long', 'lat'])
```
